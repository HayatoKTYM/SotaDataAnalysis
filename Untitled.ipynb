```python
with open("wakati.txt") as f:
    feature = [line.strip().split() for line in f.readlines()]
    #feature = [line.strip() for line in f.readlines()]
c = 1
keyword = dict()
id2word = dict()
for i in range(len(feature)):
    for j in range(len(feature[i])):
        if feature[i][j] in keyword:
            feature[i][j] = keyword[feature[i][j]]
            continue
        keyword[feature[i][j]] = c
        id2word[c] = feature[i][j]
        feature[i][j] = c
        c += 1
    feature[i] = np.array(feature[i])
id2word
```

```python
np.array(feature)
```

```python
from keras.preprocessing import sequence

feature_ = sequence.pad_sequences(feature,maxlen=29,value=0)
np.shape(feature_)
```

```python
import gensim
#model = gensim.models.KeyedVectors.load_word2vec_format('/mnt/aoni02/katayama/model.vec', binary=False)
model = gensim.models.KeyedVectors.load_word2vec_format('/Users/hayato/Desktop/fasttext/model.vec', binary=False)
zerovector = np.zeros(300)
zerovector
```

```python
feature_
```

```python
X = []#*len(feature_)
for i in range(2613):
    X.append([])
    for j in range(29):
        if feature_[i][j] == 0:
            X[i].append(zerovector)
        else:
            try:
                X[i].append(model[id2word[feature_[i][j]]])
            except:
                print(id2word[feature_[i][j]])
                X[i].append(zerovector)
```

```python
np.shape(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, label,test_size=0.2)
```

```python
from keras.models import Sequential, Model
from keras.layers.core import Dense, Activation,Dropout,Flatten
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping, TensorBoard
from keras.layers import Input,Dense,LSTM,concatenate,BatchNormalization,Bidirectional

inputs = Input(shape=(29,300))
lstm = Bidirectional(LSTM(32))(inputs)
prediction = Dense(1,activation='sigmoid')(lstm)
model = Model(inputs=inputs,outputs=prediction)
model.summary()
optimizer = Adam(lr=1e-4,beta_1=0.9,beta_2=0.999)#lr = 0.001
model.compile(loss='binary_crossentropy', optimizer=optimizer,  metrics=['accuracy'],)
early_stopping = EarlyStopping(monitor='val_loss', patience=10, verbose=1)
model.fit(X_train, y_train,
         epochs = 100,
         batch_size=128,
         validation_split=0.2,
         callbacks=[early_stopping])
```

```python
model.evaluate(X_test,y_test)
```
